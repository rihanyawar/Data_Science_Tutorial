```python
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
```

```python
tips= px.data.tips()
iris= px.data.iris()
gap= px.data.gapminder()
```

```python
iris
```

```python
gap
```

```python
temp_df= gap[gap['year']==2007]
```

```python
trace1= go.Scatter(x=temp_df['lifeExp'], y=temp_df['gdpPercap'])
trace2= go.Scatter(x=[0,1,2], y=[0,7,3], mode='lines')

data=[trace1, trace2]

layout= go.Layout(title='Life Exp vs GDP per Capita', xaxis={'title':'Life Exp'}, yaxis={'title':'GDP'})

fig= go.Figure(data, layout)
fig.show()
```

```python
px.scatter(temp_df, x='lifeExp', y='gdpPercap', color='continent', size='pop', size_max=100, hover_name='country')
```

```python
px.scatter(gap, x='lifeExp', 
           y='gdpPercap', 
           color='continent', 
           size='pop', 
           size_max=100, 
           hover_name='country',
           range_x=[30,95],
          animation_frame='year', animation_group='country')
```

```python
#Line Plot

df1= gap[gap['country'].isin(['India','Pakistan','China'])].pivot(index='year', columns='country', values='lifeExp')
```

```python
df1
```

```python
px.line(df1, x=df1.index, y= df1.columns)
```

```python
#Bar Chart
temp_df= gap[gap['country']=='India']
temp_df
```

```python
px.bar(temp_df, x='year', y='pop')
```

```python
#Grouped bar chart
temp_df=  gap[gap['country'].isin(['India','Pakistan','China'])].pivot(index='year', columns='country', values='lifeExp')
px.bar(temp_df, x=temp_df.index, y=temp_df.columns, barmode='group')
```

```python
#Histogram

temp_df = gap[gap['year']== 2007]
px.histogram(temp_df, x='lifeExp', nbins=10)
```

```python
#Pie chart 
temp_df = gap[(gap['year']==2007) & (gap['continent']=='Europe')]
px.pie(temp_df, values='pop', names='country')
```

```python
temp_df = gap[gap['year']==2007]
px.sunburst(temp_df, path=['continent', 'country'], values='pop')
```

```python
temp_df = gap[gap['year']==2007]
px.treemap(temp_df, path=[px.Constant('World'), 'continent', 'country'], values='pop', color='lifeExp')
```
