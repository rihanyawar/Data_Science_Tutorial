```python
df= pd.read_csv("DATASET/batsman_season_record.csv")
```

```python
df
```

```python
plt.bar(df['batsman'], df['2015'])
```

```python
x = np.arange(len(df['batsman']))   # positions
width = 0.35                    # bar width

# Plot bars
plt.bar(x - width/2, df['2015'], width, label='2015')
plt.bar(x + width/2, df['2016'], width, label='2016')

plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Comparison (2015 vs 2016)')
plt.xticks(x, df['batsman'])
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

companies = ['A', 'B', 'C', 'D']
sales_2015 = [20, 35, 30, 25]
sales_2016 = [25, 32, 34, 20]

x = np.arange(len(companies))

plt.bar(x, sales_2015, label='2015')
plt.bar(x, sales_2016, bottom=sales_2015, label='2016')

plt.xticks(x, companies)
plt.xlabel('Company')
plt.ylabel('Sales')
plt.title('Company-wise Sales (Stacked)')
plt.legend()

plt.show()
```
